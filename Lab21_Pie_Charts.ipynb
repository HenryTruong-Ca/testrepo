```python
import pandas as pd
import matplotlib.pyplot as plt

def load_and_preprocess_data(file_path):
    """Loads the data and preprocesses relevant columns."""
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None

def create_pie_chart(data, title, top_n=5):
    """Creates a pie chart for the given data."""

    top_values = data.value_counts().nlargest(top_n)
    plt.figure(figsize=(8, 8))
    plt.pie(top_values, labels=top_values.index, autopct='%1.1f%%', startangle=90)
    plt.title(title)
    plt.show()

# Load and preprocess the data
file_path = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = load_and_preprocess_data(file_path)

if df is not None:  # Proceed only if data loading was successful

    # Task 1: Visualizing Data Composition
    create_pie_chart(df['DatabaseWantToWorkWith'].str.split(';', expand=True).stack(), 'Top 5 Databases Respondents Want to Work With')
    create_pie_chart(df['DevType'].str.split(';', expand=True).stack(),'Top 5 Developer Types') #Added split to only count unique types.
    create_pie_chart(df['OpSysProfessional use'], 'Top Operating Systems (Professional Use)')



    # Task 2: Additional Visualizations
    create_pie_chart(df['LanguageHaveWorkedWith'].str.split(';', expand=True).stack(), 'Top 5 Programming Languages Respondents Have Worked With') # Added split and stack
    create_pie_chart(df['NEWCollabToolsHaveWorkedWith'].str.split(';', expand=True).stack(),'Top Collaboration Tools Used Professionally')#Added split to only count unique tools.


    # Task 3: Analyzing and Interpreting Composition
    create_pie_chart(df['LanguageAdmired'].str.split(';', expand=True).stack(),'Top 5 Most Admired Programming Languages')#Added split to only count unique admiration.
    create_pie_chart(df['AIToolCurrently Using'].str.split(';', expand=True).stack(),'Top 5 Tools Used for AI Development')#Added split to only count unique tools.
    create_pie_chart(df['WebframeWantToWorkWith'].str.split(';', expand=True).stack(),'Top 5 Preferred Web Frameworks')#Added split to only count unique frameworks.
    create_pie_chart(df['EmbeddedWantToWorkWith'].str.split(';', expand=True).stack(),'Top 5 Most Desired Embedded Technologies')#Added split to only count unique technologies.
```
